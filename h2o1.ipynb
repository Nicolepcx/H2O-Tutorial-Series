```python

import h2o
```

```python
##Initialize H2o cluster.If no address is mentioned than h2o will initialize a cluster on your local machine.
### To customize cluster resources refer http://docs.h2o.ai/h2o/latest-stable/h2o-r/docs/reference/h2o.init.html
h2o.init()
```

```python
### df is a h2o dataframe 


from h2o.estimators.gbm import H2OGradientBoostingEstimator

df = h2o.import_file(path="/Users/shivayogi/Downloads/safe driver prediction/train.csv")
```

```python
df.head()
```

```python
###Our dataset has 595212 rows and 59 columns . The variable target is our response variable
print (df.dim)
```

```python
response = "target"

## the response variable is an integer, we will turn it into a categorical/factor for binary classification
df[response] = df[response].asfactor()           

## use all other columns (except for the name & the response column ("survived")) as predictors
predictors = df.columns
```

```python
### Creating training , testing and validation data frames

train, valid, test = df.split_frame(
    ratios=[0.6,0.2], 
    seed=1234, 
    destination_frames=['train.hex','valid.hex','test.hex']
)
```

```python

### Create Base Models fror gradient boosting machine. 
##We will discuss more about this algorithm in detail in later tutorials
##  We will tune this algorithm in later to achieve it optimal performance

gbm = H2OGradientBoostingEstimator()
gbm.train(x=predictors, y=response, training_frame=train)

## Show a detailed model summary
print (gbm)
```

```python
#### As noted from the results below the classification performance of our algorithm is currently average.
## 

perf = gbm.model_performance(test)
perf.auc
```

```python
print (perf.auc)
```

```python
#### Get predictions for actual test set in the competition
test_kag= h2o.import_file(path="/Users/shivayogi/Downloads/safe driver prediction/test.csv")
preds_gbm = gbm.predict(test_kag)
```

```python
# Import H2O GLM:
from h2o.estimators.glm import H2OGeneralizedLinearEstimator


glm_fit1 = H2OGeneralizedLinearEstimator(family='binomial', model_id='glm_fit1')
```

```python
glm_fit1.train(x=predictors, y=response, training_frame=train)
```

```python
glm_perf1 = glm_fit1.model_performance(test)
glm_perf1
```

```python
#obtain predictions for glm
preds_glm=glm.predict(test_kag)
```

```python
###Obtain these predictions for scoring on leaderboard at https://www.kaggle.com/c/porto-seguro-safe-driver-prediction/leaderboard
pred_gbm = preds_gbm.as_data_frame()
pred_glm = preds_glm.as_data_frame()

pred_gbm.to_csv('gbm.csv', header=True, index=False)
pred_glm.tocsv('glm.csv',header=True,index=False  )
```
